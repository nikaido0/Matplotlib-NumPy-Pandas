```python
import matplotlib.pyplot as plt
import random
from matplotlib_inline import backend_inline

# 显示高清图像
backend_inline.set_matplotlib_formats('svg')

# 1、准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

# 中文显示问题
plt.rcParams['axes.unicode_minus'] = False  #用来正常显示负号

# 2、创建画布
plt.figure()

# 3、绘制图像
plt.plot(x, y_shanghai, color="r", linestyle='--', label="shanghai")
plt.plot(x, y_beijing, color="b", label="beijing")

# 显示图例
plt.legend()

# 修改x y刻度
x_label = ["11minutes{}seconds".format(i) for i in x]
plt.xticks(x[::5], x_label[::5])
plt.yticks(range(0, 40, 5))

# 显示网格
plt.grid(linestyle='--', alpha=0.5)

# 添加描述 标题
plt.xlabel("time")
plt.ylabel("temperature")
plt.title("Diagram of the temperature change between 11:00 and 12:00 in Shanghai and Beijing")

# 4、显示图像
plt.show()
```
